```python
import matplotlib.pyplot as plt
import netCDF4 as nc
%matplotlib inline
```

```python
grid = nc.Dataset('https://salishsea.eos.ubc.ca/erddap/griddap/ubcSSnBathymetryV17-02')
```

# By grid cell

```python
plt.pcolormesh(grid.variables['bathymetry'][:])
plt.colorbar()
```

# By lat/lon

```python
plt.pcolormesh(grid.variables['longitude'][:], grid.variables['latitude'][:], 
              grid.variables['bathymetry'][:])
plt.colorbar()
```
