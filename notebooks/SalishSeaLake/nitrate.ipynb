```python
import pandas as pd
import netCDF4 as nc
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
from salishsea_tools import viz_tools, nc_tools
import glob
import datetime
%matplotlib inline
```

```python
apr17 = sorted(glob.glob('/results/SalishSea/nowcast-green/*apr17/*1h*ptrc*'))
```

```python
with nc_tools.scDataset(apr17) as f:
    surface = f.variables['nitrate'][:,0,625, 165]
    depth5 = f.variables['nitrate'][:,4,625, 165]
```

```python
mesh = nc.Dataset('/data/vdo/MEOPAR/NEMO-forcing/grid/mesh_mask201702.nc')
```

```python
fig, ax = plt.subplots(figsize = ((10,10)))
ax.pcolormesh(mesh.variables['tmask'][0,0,...])
viz_tools.set_aspect(ax)
ax.plot(165, 625, 'ro')
```

```python
dates = np.array([datetime.datetime(2017,4,1) + datetime.timedelta(hours = n) 
                  for n in range(720)])
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
ax.plot(dates, surface, label = 'surface')
ax.plot(dates, depth5, label = '5m')
ax.legend()
ax.grid('on')
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
ax.plot(dates[:100], surface[:100], label = 'surface')
ax.plot(dates[:100], depth5[:100], label = '5m')
ax.legend()
ax.grid('on')
```
