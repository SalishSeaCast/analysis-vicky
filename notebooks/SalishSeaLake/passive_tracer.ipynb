Tracking a passive tracer in SalishSeaLake

```python
import netCDF4 as nc
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Starting Density

```python
f = nc.Dataset('/data/vdo/MEOPAR/SalishSea/gradienttest/SalishSea_1h_20160601_20160601_ptrc_T.nc')
mesh = nc.Dataset('/data/vdo/MEOPAR/NEMO-forcing/grid/lake_meshmask201702.nc')
bathy = nc.Dataset('/data/vdo/MEOPAR/NEMO-forcing/grid/lake_bathy201702.nc')
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.pcolormesh(bathy.variables['Bathymetry'][:])
ax.plot((0,284), (200,200), 'r-')
```

```python
fig, ax = plt.subplots(figsize = (8,8))
den = ax.pcolormesh(np.ma.masked_array(f.variables['mytracer'][0,:,200, :],
                                       mask = 1 - mesh.variables['tmask'][0,:,200,:]))
fig.colorbar(den, ax=ax)
ax.set_ylim(40,0)
```
