Some Nitrate-Diatom plots for Deep Bay

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import glob
from salishsea_tools import nc_tools 
import netCDF4 as nc
```

```python
Jan = sorted(glob.glob('/results/SalishSea/hindcast/*jan15/*1h*ptrc_T*'))
Feb = sorted(glob.glob('/results/SalishSea/hindcast/*feb15/*1h*ptrc_T*'))
Mar = sorted(glob.glob('/results/SalishSea/hindcast/*mar15/*1h*ptrc_T*'))
Apr = sorted(glob.glob('/results/SalishSea/hindcast/*apr15/*1h*ptrc_T*'))
May = sorted(glob.glob('/results/SalishSea/hindcast/*may15/*1h*ptrc_T*'))
Jun = sorted(glob.glob('/results/SalishSea/hindcast/*jun15/*1h*ptrc_T*'))
Jul = sorted(glob.glob('/results/SalishSea/hindcast/*jul15/*1h*ptrc_T*'))
Aug = sorted(glob.glob('/results/SalishSea/hindcast/*aug15/*1h*ptrc_T*'))
Sep = sorted(glob.glob('/results/SalishSea/hindcast/*sep15/*1h*ptrc_T*'))
Oct = sorted(glob.glob('/results/SalishSea/hindcast/*sep15/*1h*ptrc_T*'))
Nov = sorted(glob.glob('/results/SalishSea/hindcast/*nov15/*1h*ptrc_T*'))
Dec = sorted(glob.glob('/results/SalishSea/hindcast/*dec15/*1h*ptrc_T*'))
```

```python
with nc_tools.scDataset(Oct) as f:
    t = f.variables['nitrate'][:, :7, 599, 125]
    s = f.variables['diatoms'][:, :7, 599, 125]
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12,12))
    ax1.plot(t[0,0],s[0,0],'y*', markersize = 20)
    ax1.plot(t[:,0],s[:,0])
    ax1.grid('on')
    ax1.set_ylabel('Diatoms')
    ax1.set_xlabel('Nitrate')
    ax1.set_title(' depth= 0')
    ax2.plot(t[0,-1],s[0,-1],'y*', markersize = 20)
    ax2.plot(t[:,-1],s[:,-1])
    ax2.grid('on')
    ax2.set_ylabel('Diatoms')
    ax2.set_xlabel('Nitrate')
    ax2.set_title('depth layer 6')
    #fig.suptitle('Nitrate-Diatom plot at lon=' + str(lon) + ', lat=' + str(lat) + ',from hour=' + str(start_time) 
                 #+ ' to hour= ' + str(end_time -1), fontsize = 20)
    plt.savefig('Nitrate-Diatoms-Oct.png', bbox_inches='tight')
```
