```python
import numpy as np
import xarray as xr
import netCDF4 as nc
import matplotlib.pyplot as plt
import scipy.io as sio
import pandas as pd
import datetime
import xarray as xr
from salishsea_tools import tidetools, geo_tools, viz_tools, nc_tools
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.colors import LogNorm
import glob
import pickle
import matplotlib.lines as mlines
%matplotlib inline
```

```python
x = np.arange(0,100)
```

```python
def his(ni_path, si_path):
    si = pickle.load(open(si_path, 'rb'))
    ni = pickle.load(open(ni_path, 'rb'))
    fig, ax = plt.subplots(figsize = (8,8))
    c, xedge, yedge, im = ax.hist2d(ni, si, bins = 100, norm=LogNorm())
    fig.colorbar(im, ax=ax)
    ax.set_xlabel('N')
    ax.set_ylabel('Si')
    ax.set_title('N V Si, nowcast-green 2015')
    ax.plot(x, x, 'r-', label = 'slope = 1')
    ax.plot(x, 2*x, 'g-', label = 'slope = 2')
    ax.legend()
    ax.set_ylim(0, si.max()+5)
    ax.set_xlim(0, ni.max()+5)
    return ax
```
