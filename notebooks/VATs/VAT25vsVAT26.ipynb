```python
import numpy as np
import xarray as xr
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
VAT25T = xr.open_dataset('/data/vdo/MEOPAR/completed-runs/VAT25/SalishSea_1h_20150507_20150509_grid_T_20150508-20150508.nc')
VAT26T = xr.open_dataset('/data/vdo/MEOPAR/completed-runs/VAT26/SalishSea_1h_20150507_20150509_grid_T_20150508-20150508.nc')
VAT25V = xr.open_dataset('/data/vdo/MEOPAR/completed-runs/VAT25/SalishSea_1h_20150507_20150509_grid_V_20150508-20150508.nc')
VAT26V = xr.open_dataset('/data/vdo/MEOPAR/completed-runs/VAT26/SalishSea_1h_20150507_20150509_grid_V_20150508-20150508.nc')
```

```python
figs, (ax1,ax2) = plt.subplots(1,2,figsize=(10,8))
VAT25T.sossheig.isel(time_counter=23).plot(ax=ax1,vmin = -3, vmax = 3,cmap = 'bwr')
VAT26T.sossheig.isel(time_counter=23).plot(ax=ax2, vmin = -3, vmax = 3, cmap = 'bwr')
```

```python
(VAT26T.sossheig.isel(time_counter=23) - VAT25T.sossheig.isel(time_counter=23)).plot(figsize=(10,8),
                                                                                    vmin = -0.02,
                                                                                    vmax = 0.02,
                                                                                    cmap = 'bwr')
```

```python
figs, (ax1,ax2) = plt.subplots(1,2,figsize=(10,8))
VAT25T.vosaline.isel(time_counter=23,deptht = 0).plot(ax=ax1)
VAT26T.vosaline.isel(time_counter=23,deptht=0).plot(ax=ax2)
```

```python
(VAT26T.vosaline.isel(time_counter=23,deptht=0) - VAT25T.vosaline.isel(time_counter=23,deptht = 0)).plot(figsize=((10,8)))
```
