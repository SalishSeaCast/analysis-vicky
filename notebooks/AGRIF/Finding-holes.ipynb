```python
import numpy as np
import netCDF4 as nc
import xarray as xr
import matplotlib.pyplot as plt
import numpy.ma as ma
%matplotlib inline
```

```python
runoff_init = nc.Dataset('/data/vdo/MEOPAR/completed-runs/runoff/1_output.init.nc')
runoff_mask = nc.Dataset('/data/vdo/MEOPAR/completed-runs/runoff/1_mesh_mask.nc')
agrif_init = nc.Dataset('/data/vdo/MEOPAR/completed-runs/changing_agrif/1_output.init.nc')
agrif_mask = nc.Dataset('/data/vdo/MEOPAR/completed-runs/changing_agrif/1_mesh_mask.nc')
vis_init = nc.Dataset('/data/vdo/MEOPAR/completed-runs/viscosity/1_output.init.nc')
vis_mask = nc.Dataset('/data/vdo/MEOPAR/completed-runs/viscosity/1_mesh_mask.nc')
vis2_init = nc.Dataset('/data/vdo/MEOPAR/completed-runs/viscosity2/1_output.init.nc')
vis2_mask = nc.Dataset('/data/vdo/MEOPAR/completed-runs/viscosity2/1_mesh_mask.nc')
```

```python
fig,ax = plt.subplots(1,1,figsize=(14,12))
s = ma.masked_array(vis_init.variables['sossheig'][0,:,:], mask = 1 - vis_mask.variables['tmask'][0,0,:,:])
mesh = ax.pcolormesh(s,#vmin = 15, vmax = 33
                     )
fig.colorbar(mesh)
plt.title('Changing the Viscosity')
```

```python
fig,ax = plt.subplots(1,1,figsize=(14,12))
s = ma.masked_array(vis_init.variables['vosaline'][0,0,:,:], mask = 1 - vis_mask.variables['tmask'][0,0,:,:])
mesh = ax.pcolormesh(s,#vmin = 0, vmax = 33
                     )
fig.colorbar(mesh)
plt.title('Surface Salinity While Changing the Viscosity')
```
