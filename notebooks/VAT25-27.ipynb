```python
import xarray as xr
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
VAT25 = xr.open_dataset('/data/vdo/MEOPAR/completed-runs/VAT25/SalishSea_1h_20150507_20150509_grid_T_20150508-20150508.nc')
VAT27 = xr.open_dataset('/data/vdo/MEOPAR/completed-runs/VAT27/SalishSea_1h_20150507_20150509_grid_T_20150508-20150508.nc')
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=((10,8)))
VAT25.sossheig.isel(time_counter=23).plot(ax=ax1,vmin = -3, vmax=3, cmap = 'bwr')
VAT27.sossheig.isel(time_counter=23).plot(ax=ax2,vmin = -3, vmax=3, cmap = 'bwr')
```

```python
(VAT25.sossheig.isel(time_counter=23) - VAT27.sossheig.isel(time_counter=23)).plot(figsize=((10,8)),
                                                                                  vmin = -0.02,
                                                                                  vmax = 0.02,
                                                                                  cmap = 'bwr')
```
