Looking at the output.abort files from running with patch

```python
import netCDF4 as nc
import matplotlib.pyplot as plt
import numpy.ma as ma
%matplotlib inline
```

```python
mask = nc.Dataset('/data/vdo/MEOPAR/NEMO-forcing/grid/mesh_mask201702.nc')
abort = nc.Dataset('/ocean/vdo/MEOPAR/completed-runs/output.abort.nc')
```

```python
fig,ax = plt.subplots(1,1,figsize = ((10,8)))
ssh = ma.masked_array(abort.variables['sossheig'][0,:,:], mask = 1 - mask.variables['tmask'][0,0,:,:])
mesh = ax.pcolormesh(ssh, vmin = -0.1, vmax = 0.1, cmap = 'coolwarm')
fig.colorbar(mesh)
plt.plot(127,733,'go')
plt.title('SSH')
```

```python
fig,ax = plt.subplots(1,1,figsize = ((10,8)))
ssh = ma.masked_array(abort.variables['vosaline'][0,0,:,:], mask = 1 - mask.variables['tmask'][0,0,:,:])
mesh = ax.pcolormesh(ssh, vmin = 0, vmax = 35
                    )
fig.colorbar(mesh)
plt.plot(127,733,'ro')
plt.title('Salinity')
```
