```python
f0 = nc.Dataset('/ocean/vdo/MIDOSS/Lagrangian_AKNS_crude-0_AKNS-InOneSSGrid-0.nc')
f1 = nc.Dataset('/ocean/vdo/MIDOSS/Lagrangian_AKNS_crude-1_AKNS-InOneSSGrid-1.nc')
f2 = nc.Dataset('/ocean/vdo/MIDOSS/Lagrangian_AKNS_crude-2_AKNS-InOneSSGrid-2.nc')
f3 = nc.Dataset('/ocean/vdo/MIDOSS/Lagrangian_AKNS_crude-3_AKNS-InOneSSGrid-3.nc')
f4 = nc.Dataset('/ocean/vdo/MIDOSS/Lagrangian_AKNS_crude-4_AKNS-InOneSSGrid-4.nc')
```

## top left - bottom right

```python
for n in np.arange(1,21):
    fig, ax = plt.subplots()
    ax.set_title('h = ' +str(n))
    mesh = ax.pcolormesh(np.sum(f0['OilConcentration_3D'][n,:,250:360, 200:270], axis=0) - 
                         np.sum(f4['OilConcentration_3D'][n,:,250:360, 200:270], axis=0),
                         cmap = 'bwr', vmin = -0.3, vmax = 0.3)
    fig.colorbar(mesh, ax=ax)
```
