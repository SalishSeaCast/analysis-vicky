```python
import netCDF4 as nc
import numpy as np 
import matplotlib.pyplot as plt
from salishsea_tools import viz_tools
%matplotlib inline
```

```python
bathy = nc.Dataset('/data/vdo/MEOPAR/grid/bathymetry_201702.nc')
```

```python
import cmocean
```

# Vertically

```python
fig, ax = plt.subplots(figsize = (12,12))
b = ax.pcolormesh((bathy.variables['Bathymetry'][:-1,:] - bathy.variables['Bathymetry'][1:, :]) /
               ((bathy.variables['Bathymetry'][:-1, :] + bathy.variables['Bathymetry'][1:,:])/2),
                 cmap = cmocean.cm.oxy)
fig.colorbar(b, ax=ax)
viz_tools.set_aspect(ax)
ax.set_xlim(100, 200)
ax.set_ylim(0,200)
```

# Horizontally 

```python
fig, ax = plt.subplots(figsize = (12,12))
b = ax.pcolormesh((bathy.variables['Bathymetry'][:,:-1] - bathy.variables['Bathymetry'][:, 1:]) /
               ((bathy.variables['Bathymetry'][:, :-1] + bathy.variables['Bathymetry'][:,1:])/2),
                 cmap = cmocean.cm.oxy)
fig.colorbar(b, ax=ax)
viz_tools.set_aspect(ax)
ax.set_xlim(100, 200)
ax.set_ylim(0,200)
```
