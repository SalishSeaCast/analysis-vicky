## surface

```python
for n in np.arange(1,21):
    fig, ax = plt.subplots()
    ax.set_title('h = ' +str(n))
    mesh = ax.pcolormesh(f['OilConcentration_3D'][n,39,200:300, 200:350], 
                         cmap = 'Blues_r', vmin = 0, vmax = 0.5)
    fig.colorbar(mesh, ax=ax)
```

## sum of depth !=0

```python
for n in np.arange(1,21):
    fig, ax = plt.subplots()
    ax.set_title('h = ' +str(n))
    mesh = ax.pcolormesh(np.sum(f['OilConcentration_3D'][n,:39,200:300, 200:350], axis = 0), 
                  cmap = 'Blues_r', vmin = 0, vmax = 0.3)
    fig.colorbar(mesh, ax=ax)
```
